# **Data Preprocessing**

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
X.head()
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
```

```python
numeric_cols = [column for column in X_train.columns if X_train[column].dtype == 'float64']
print('Numerical columns:', numeric_cols)
categorical_cols = [column for column in X_train.columns if X_train[column].dtype == 'object']
print('Categorical columns:', categorical_cols)
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

# **Model Building**

```python
param_grid = {'C' : [0.1,1,10],'gamma' : [1,0.1,0.01], 'kernel' : ['rbf'], 'probability' : [True]}
gs = GridSearchCV(SVC(),param_grid,cv=5)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_params_
```

```python
gs.best_estimator_
```

```python
y_pred = gs.predict(X_test)
```

# **Metrics and Evaluation**

```python
print('Train data score:',gs.score(X_train,y_train))
print('Test data score:',gs.score(X_test,y_test))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
pred_prob = gs.predict_proba(X_test)
```

```python
roc_auc_score(y_test, pred_prob[:,1])
```
